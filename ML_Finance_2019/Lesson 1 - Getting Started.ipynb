```python
import pandas as pd    # Import Pandas
import numpy as np     # Import Numpy, 
import seaborn as sns  # Import Seaborn, visualisation library
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

download the data and store it in the variable called `df`

```python
df = pd.read_csv('iris.data')
```

Let's review our data using the method **.head()**

```python
df.head()
```

note that the first row was used as header. We need to change that

```python
df = pd.read_csv('iris.data', header=None)
df.head()
```

Let's give our columns of data some useful names. You obtain these information from the UCI Machine Learning Repository

```python
col_name =   ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
df.columns = col_name
df.head()
```

## Iris Data from Seaborn

The dataset is also available from the Seaborn library

```python
iris = sns.load_dataset('iris')
iris.head()
```

Let's have a look at some descriptive data with the use of **.describe()** method

```python
iris.describe()
```

For some meta data information. You can use the **.info()** method

```python
iris.info()
```

Next step is to group by the different flower species and have a look at the data size.

```python
print(iris.groupby('species').size())
```

## Visualisation

```python
sns.pairplot(iris, hue="species", size=2.5);
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(10,8));
plt.show();
```

```python
plt.figure(figsize=(12,8));
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris);
```

```python
iris.boxplot(by='species', figsize=(12,8));
```

```python
sns.pairplot(iris, hue="species", diag_kind="kde");
```
